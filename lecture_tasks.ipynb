Q . give a difference between indexing and slicing 

ans : indexing accesses a single element from a sequence using a specific index, such as my_list[2], which retrieves the third item. Indexing starts from 0 for positive indices and -1 for negative ones from the end.

Slicing extracts a range of elements using start:end syntax, like my_list[1:4], creating a new subsequence


Notes----

slicing refers to the process of extracting a portion of an array or a subarray 
it allows you to access specific elements or ranges of elements in a numpy array 

indexing is the process of accessing a single element from a sequence using its position or index 

2 types of indexing :

positive : left side - 0
negative : right side - (-1)

it retrives a single elements at a specific position
